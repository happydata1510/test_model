```python
# ====== 5. Flask ì•± ìƒì„± ======
app = Flask(__name__)
CORS(app)

@app.route('/')
def home():
    return render_template_string(HTML_TEMPLATE)

@app.route('/predict', methods=['POST'])
def predict():
    try:
        # ì…ë ¥ ë°ì´í„° ë°›ê¸°
        data = request.json
        print(f"ë°›ì€ ë°ì´í„°: {data}")

        # features ë°°ì—´ ì¶”ì¶œ
        if 'features' not in data:
            return jsonify({'error': 'features ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤'}), 400

        features = data['features']

        # 8ê°œ íŠ¹ì„±ì´ ëª¨ë‘ ìˆëŠ”ì§€ í™•ì¸
        if len(features) != 8:
            return jsonify({'error': f'8ê°œì˜ íŠ¹ì„±ì´ í•„ìš”í•©ë‹ˆë‹¤. ë°›ì€ ê°œìˆ˜: {len(features)}'}), 400

        # numpy ë°°ì—´ë¡œ ë³€í™˜
        features_array = np.array(features).reshape(1, -1)
        print(f"íŠ¹ì„± ë°°ì—´ shape: {features_array.shape}")

        # ì˜ˆì¸¡
        prediction = model.predict(features_array)[0]
        print(f"ì˜ˆì¸¡ê°’: {prediction}")

        # ê²°ê³¼ë¥¼ ë‹¬ëŸ¬ë¡œ ë³€í™˜ (ì‹­ë§Œ ë‹¬ëŸ¬ â†’ ë‹¬ëŸ¬)
        price_in_dollars = int(prediction * 100000)

        # ê²°ê³¼ ë°˜í™˜
        result = {
            'price': price_in_dollars,
            'success': True
        }

        print(f"ë°˜í™˜í•  ê²°ê³¼: {result}")
        return jsonify(result)

    except Exception as e:
        print(f"ì„œë²„ ì—ëŸ¬: {str(e)}")
        import traceback
        traceback.print_exc()
        return jsonify({'error': str(e)}), 500
```
